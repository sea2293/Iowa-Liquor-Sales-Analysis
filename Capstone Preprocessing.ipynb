```python
#step 1
#lets remove 'Pack' for item consistency
products = products.drop('Pack', axis=1)
products
```

```python
#step 2
#map item to majority category for consistency

item_cat = products.groupby(['ItemNumber']).agg({'Category': 'value_counts'}).rename(columns = {'Category': 'ValueCount'})
item_cat = item_cat.reset_index()
item_cat['Rank'] = item_cat.groupby('ItemNumber')['ValueCount'].rank(ascending = False)
item_cat_unique = item_cat[item_cat['Rank']==1]
item_cat_unique
```

```python
products = pd.merge(products, item_cat_unique, how = 'left', on = 'ItemNumber')
products.head()
```

```python
#now drop excess columns
products.drop(['Category_x', 'ValueCount', 'Rank'], axis=1, inplace=True)
products.rename(columns = {'Category_y': 'Category'}, inplace=True)
products.head()
```

```python
#step 3
#map item to majority category name for consistency
cat_group = products.groupby(['Category', 'CategoryName'])['ItemNumber'].count().reset_index().rename(columns = {'ItemNumber': 'ValueCount'})
cat_group['Rank'] = cat_group.groupby('Category')['ValueCount'].rank(ascending = False)
cat_group_unique = cat_group[cat_group['Rank'] == 1]
```

```python
products = pd.merge(products, cat_group_unique, how = 'left', on = 'Category')
products
```

```python
products.drop(['CategoryName_x', 'ValueCount', 'Rank'], axis=1, inplace = True)
products.rename(columns = {'CategoryName_y': 'CategoryName'}, inplace=True )
products
```

```python
#step 4
#map item to majority Item Description for consistency
item_desc = products.groupby('ItemNumber').agg({'ItemDescription': 'value_counts'})
item_desc.rename(columns = {'ItemDescription': 'ValueCount'}, inplace=True)
item_desc.reset_index(inplace=True)
item_desc['Rank'] = item_desc.groupby('ItemNumber')['ValueCount'].rank(ascending = False)
item_desc_unique = item_desc[item_desc['Rank']==1]
```

```python
products = pd.merge(products, item_desc_unique, how = 'left', on = 'ItemNumber')
products
```

```python
products.drop(['ItemDescription_x', 'ValueCount', 'Rank'], axis=1, inplace=True)
products.rename(columns = {'ItemDescription_y': 'ItemDescription'}, inplace=True)
products
```

```python
#step 5
#map item to majority bottle volume for consistency
bottle_vol = products.groupby('ItemNumber').agg({'BottleVolume(ml)': 'value_counts'}).rename(columns = {'BottleVolume(ml)': 'ValueCount'}).reset_index()
bottle_vol['Rank'] = bottle_vol.groupby('ItemNumber')['ValueCount'].rank(ascending = False)
bottle_vol_unique = bottle_vol[bottle_vol['Rank'] == 1]
```

# Clean Price info

```python
prices = master_df[price_info]
```

```python
#rename columns
prices.columns = ['ItemNumber', 'StateBottleCost', 'StateBottleRetail', 'Date']
```

```python
prices
```

```python
#convert date column to date type
prices['Date'] = pd.to_datetime(prices['Date'], infer_datetime_format=True)
```

```python
#removes letter from item number
prices = prices.replace('x904631', '904631')
#checks to see if all value counts are still 1
prices['ItemNumber'].value_counts(dropna = False)
```

```python
#convert item numbers to integer type
prices.loc[:, 'ItemNumber'] = prices['ItemNumber'].astype(int)
```

```python
#sort by date
prices = prices.sort_values(by='Date')
```

```python
prices
```

```python
#check for NAs
prices[prices['ItemNumber'].isna()]
```

```python
prices[prices['StateBottleCost'].isna()]
```

```python
prices[prices['StateBottleRetail'].isna()]
```

```python
#figure out missing price info
#use item number and median price to fill NAs
prices['StateBottleCost'] = prices.groupby('ItemNumber')['StateBottleCost'].transform(lambda x: x.fillna(x.median()))
prices[prices['StateBottleCost'].isna()]
```

```python
prices['StateBottleRetail'] = prices.groupby('ItemNumber')['StateBottleRetail'].transform(lambda x: x.fillna(x.median()))
prices[prices['StateBottleCost'].isna()]
```

```python
prices[prices['Date'].isna()]
```

```python
#feature engineering
prices['StateBottleProfit'] = prices['StateBottleRetail'] - prices['StateBottleCost']
```

```python
prices
```

```python
product_price = prices.groupby(['ItemNumber', 'Date']).agg({'StateBottleRetail':'mean','StateBottleCost':'mean','StateBottleProfit':'mean'})
product_price.reset_index(inplace=True)
```

```python
product_price
```

```python
#export df as csv
prices.to_csv('ia_prices.csv')
```

```python
#deal with repeats for store name
store_name = stores.groupby(['StoreNumber']).agg({'StoreName': 
                                                  'value_counts'}).rename(columns = {'StoreName': 
                                                                                     'ValueCount'}).reset_index()

#keep only rank 1 stores
store_name['Rank'] = store_name.groupby('StoreNumber')['ValueCount'].rank(ascending = False)
store_name_unique = store_name.loc[store_name.Rank == 1].drop(['ValueCount', 
                                                               'Rank'], axis=1)

#drop duplicate columns
stores = stores.drop(['StoreName'], axis=1)

#merge the dfs
stores = pd.merge(stores, store_name_unique, how='left', 
                 on='StoreNumber')
stores.head()
```

```python
#deal with repeats for store number
store_loc = stores.groupby(['StoreNumber', 'County', 'CountyNumber', 'City', 'ZipCode','StoreLocation']
                          ).agg({'Address':'value_counts'}).rename(columns = {'Address':'ValueCount'}).reset_index()

#keep only rank 1 stores
store_loc['Rank'] = store_loc.groupby('StoreNumber')['ValueCount'].rank(ascending = False)

store_loc_unique = store_loc.loc[store_loc.Rank == 1].drop(['ValueCount', 
                                                            'Rank'], axis=1)

# drop columns from store
stores = stores.drop(['Address','City', 'ZipCode', 'StoreLocation', 
                    'CountyNumber', 'County'], axis=1)

#merge the dfs
stores = pd.merge(stores, store_loc_unique, how='left', 
                 on='StoreNumber')
```

```python
#rename columns
vendor.columns = ['VendorName', 'VendorNumber']
```

```python
#check for NA vendor numbers
vendor[vendor['VendorNumber'].isna()]
```

```python
#check for NA vendor names
vendor[vendor['VendorName'].isna()]
```

```python
#remove NAs
vendor = vendor[~vendor['VendorNumber'].isna()]
```

```python
#TODO
#change vendor names to lowercase
vendor['VendorName'] = vendor['VendorName'].str.lower()
```

```python
ven_name = vendor.groupby('VendorNumber').agg({'VendorName': 'value_counts'})
ven_name.rename({'VendorName':'ValueCount'},axis=1,inplace = True)
ven_name = ven_name.reset_index()
ven_name['Rank'] = ven_name.groupby('VendorNumber')['ValueCount'].rank(ascending = False)
```

```python
vendors = ven_name[ven_name['Rank'] == 1.0].drop(['ValueCount','Rank'],axis=1)
```

```python
vendors
```

```python
#change vendor number to int type
vendors.loc[:, 'VendorNumber'] = vendors['VendorNumber'].astype(int)
```

```python
#check for repeat vendor numbers
vendors['VendorNumber'].value_counts()
```

```python
#export df as csv
vendors.to_csv('ia_vendors.csv')
```
